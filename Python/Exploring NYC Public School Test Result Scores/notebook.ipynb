Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

#Counting 'best math results' treshold 
treshold = 0.8 * 800
#Selecting only 2 relevant columns and filtering 'average_math' column to be equal or over the treshold
best_math_columns = schools[["school_name","average_math"]][schools["average_math"] >= treshold]
#Sorting previously created subset by the 'average_math' in descending order and creating new dataframe 
best_math_schools = best_math_columns.sort_values("average_math", ascending=False)
#Printing the results
print(best_math_schools)

#Creating new column 'total_SAT'
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
#Creating subset containing two columns and 10 top values based on 'total_SAT'
top_10_schools = schools[["school_name", "total_SAT"]].nlargest(10, "total_SAT")
print(top_10_schools)

#Aggregating data by grouping them by "borough" column and rounding results to the second decimal.
aggregated_data = schools.groupby("borough").agg(
    num_schools=("school_name", "count"),
    average_SAT=("total_SAT", "mean"),
    std_SAT=("total_SAT", "std")
).round(2)

#Sorting the results by "std_SAT" column and showing only 1 row.
largest_std_dev = aggregated_data.sort_values(by="std_SAT", ascending=False).head(1)

print(largest_std_dev)


```
